## Задание 1
**(0.25 балла)** Напишите свою версию версию функции активации [ReLU](https://pytorch.org/docs/stable/generated/torch.nn.ReLU.html#torch.nn.ReLU)




````python
# Из ячейки не нужно убирать или добавлять другие импорты или объединять с тдругими ячейками
# Это может сломать автотесты

import torch
import torch.nn as nn

class MyReLU(nn.Module):

    def forward(self, input):
        # hint! Входной тензор нужно скопировать,
        # чтобы действиями внутри этого метода не привели к изменению внешнего аргумента
        input_clone = torch.clone(input)

        #---------------------------------------------------
        #все это происходит ПОСЛЕ ЛИНЕЙНОГО ПРЕОБРАЗОВАНИЯ! (функция только на активацию!)
        #---------------------------------------------------

        #применяем ReLU: заменяем все отрицательные значения на 0

        # torch.clamp ограничивает значения снизу (min=0), оставляя положительные без изменений (torch.clamp(x, min=0) эквивалентен max(0, x) поэлементно.)

        output = torch.clamp(input_clone, min=0.0)
        # возвращаем результат — тензор той же формы, что и input, но с ReLU-преобразованием
        return output
````

## Задание 2
**(0.25 балла)** Напишите свою версию версию функции активации [LeakyReLU](https://pytorch.org/docs/stable/generated/torch.nn.LeakyReLU.html#torch.nn.LeakyReLU)




````python
# Из ячейки не нужно убирать или добавлять другие импорты или объединять с тдругими ячейками
# Это может сломать автотесты

import torch
import torch.nn as nn

class MyLeakyReLU(nn.Module):

    def __init__(self, negative_slope=0.01):

        #---------------------------------------------------
        # Когда будем создавать экземпляр класса MyLeakyReLU(negative_slope = x) , надо будет указать аргумент, которые задали в __init_ как параметр
        # self связывает метод(init вызывается автоматом при создании экземпляра) с классом
        # внутри фукнции init задаются свойства, атрибуты, которые будут присвоены экземпляру класса
        #---------------------------------------------------

        # Вызываем конструктор родительского класса nn.Module
        super().__init__()
        # Сохраняем параметр наклона для отрицательных значений
        # Значение по умолчанию 0.01 совпадает с реализацией в PyTorch
        self.negative_slope = negative_slope #инициализурем атрибут создаваемых экземлпяров

    def forward(self, input):
        # hint! Входной тензор нужно скопировать,
        # чтобы действиями внутри этого метода не привели к изменению внешнего аргумента
        input_clone = torch.clone(input)
        # Создаём маску: True где элементы положительные, False где отрицательные
        positive_mask = input_clone > 0
        # Применяем LeakyReLU:
        # - где positive_mask=True: оставляем значение как есть (x)
        # - где positive_mask=False: умножаем на negative_slope (negative_slope * x)
        output = torch.where(positive_mask, input_clone, self.negative_slope * input_clone)
        # Возвращаем результат — тензор той же формы, что и input
        return output
````

## Задание 3
**(0.25 балла)** Напишите свою версию версию функции активации [Sigmoid](https://pytorch.org/docs/stable/generated/torch.nn.Sigmoid.html#torch.nn.Sigmoid)



````python
# Из ячейки не нужно убирать или добавлять другие импорты или объединять с тдругими ячейками
# Это может сломать автотесты

import torch
import torch.nn as nn

class MySigmoid(nn.Module):

    def forward(self, input):
        # для этого класса копировать входной тензор уже нет необходимости, почему?

        # Потому что операции torch.exp и деление создают НОВЫЙ тензор,
        # а не модифицируют исходный input на месте (in-place).
        # Все операции здесь — out-of-place, поэтому внешний тензор остаётся неизменным.

                # Вычисляем сигмоиду по формуле: σ(x) = 1 / (1 + exp(-x))
        # torch.exp(-input) — вычисляет экспоненту от -input
        # Результат — новый тензор, исходный input не затрагивается
        output = 1.0 / (1.0 + torch.exp(-input))

        # Возвращаем результат — тензор той же формы, что и input
        return output

````

## Задание 4
**(0.25 балла)** Напишите свою версию версию функции активации [ELU](https://pytorch.org/docs/stable/generated/torch.nn.ELU.html#torch.nn.ELU)




````python
# Из ячейки не нужно убирать или добавлять другие импорты или объединять с тдругими ячейками
# Это может сломать автотесты

import torch
import torch.nn as nn

class MyELU(nn.Module):

    def __init__(self, alpha=1.0): #Инициализируем модуль(метод) с параметром alpha.

        #---------------------------------------------------
        # Когда будем создавать экземпляр класса MyELU(alpha = x) , надо будет указать аргумент, которые задали в __init_ как параметр
        # self связывает метод(init вызывается автоматом при создании экземпляра) с классом
        # внутри фукнции init задаются свойства, атрибуты, которые будут присвоены экземпляру класса
        #---------------------------------------------------

        # Вызываем конструктор родительского класса nn.Module
        super().__init__()
        # Сохраняем параметр alpha с значением по умолчанию 1.0,
        # что совпадает со стандартной реализацией PyTorch
        self.alpha = alpha #инициализурем атрибут создаваемых экземлпяров

    def forward(self, input):
        input_clone = torch.clone(input)
        # Создаём булеву маску: True для положительных значений (x > 0),
        # False для неположительных (x <= 0)
        positive_mask = input_clone > 0

        # Для отрицательных значений вычисляем alpha * (exp(x) - 1)
        # torch.exp(input_clone) - 1.0 — поэлементное вычисление e^x - 1
        negative_part = self.alpha * (torch.exp(input_clone) - 1.0)

        # Применяем условное присваивание:
        # - где positive_mask=True: оставляем исходное значение input_clone
        # - где positive_mask=False: используем negative_part
        output = torch.where(positive_mask, input_clone, negative_part)

        # Возвращаем результат — тензор той же формы, что и input
        return output
````

## Тест


````python
import pytest
import torch
import torch.nn as nn

def _test_activation(myactivation, torch_activation):
    print(myactivation)

    with torch.no_grad():

        randinput = torch.rand([100])
        myactivation_output = myactivation(randinput)

        assert id(myactivation_output) != id(randinput), 'pytorch activation function must return new tensor'

        for _ in range(100):
            randinput = torch.rand([5, 5, 5])

            assert torch.allclose(myactivation(randinput), torch_activation(randinput)), 'activation output is not equals to touch ones output'

def test_relu():

    my_activation = MyReLU()

    _test_activation(my_activation, nn.ReLU())

def test_leaky_relu():

    my_activation = MyLeakyReLU()

    _test_activation(my_activation, nn.LeakyReLU())

def test_sigmoid():

    my_activation = MySigmoid()

    _test_activation(my_activation, nn.Sigmoid())

def test_elu():

    my_activation = MyELU()

    _test_activation(my_activation, nn.ELU())


test_relu()
test_leaky_relu()
test_sigmoid()
test_elu()
````
